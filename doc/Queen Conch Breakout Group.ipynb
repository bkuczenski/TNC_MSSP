```python
queen_conch = FisheryGuide(E)
```

```python
def save_progress():
    qc_answers = queen_conch.save_answers()
    MSSP.write_to_json(qc_answers, outdir=gdir, outfile=ansfile)
    
def load_answers():
    answers_in = MSSP.read_json(os.path.join(gdir, ansfile))
    queen_conch.load_answers(answers_in)
    
def save_engine():
    engine_out = E.serialize()
    MSSP.write_to_json(engine_out, gdir)
```
