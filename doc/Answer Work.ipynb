```python
"""
manually verify that all satisfied_by questions have the same set of valid answers as the 
questions that satisfy them.
"""

sats = [k for k, v in enumerate(E._questions) if v is not None and len(v.satisfied_by) > 0]

for i in sats:
    
    qs = E._questions[i].satisfied_by
    print "Satisfier answers:"
    for q in qs:
        print E._questions[q].valid_answers
    print "Satisfied answers:"
    print E._questions[i].valid_answers
    print '\n'
```
